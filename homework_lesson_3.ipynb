### Задание 3

1) Напишите код, который будет переводить полярные координаты в декартовы.

2) Напишите код, который будет рисовать график окружности в полярных координатах.

3) Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.


```python
def from_polar_to_cartesian(r, rad):
    x = r*np.cos(rad)
    y = r*np.sin(rad)
    return x, y
```

```python
from_polar_to_cartesian(5, np.pi)
```

```python
def polar_circle(R):
    x = np.linspace(0, np.pi, 1000)
    y = 2*R*np.sin(x)
    plt.polar(x, y)
    
polar_circle(7)
```

```python
def polar_line(r, a):
    x = np.linspace(0, 1/3*np.pi, 500)
    y = r / np.cos(x - a)
    plt.polar(x, y)
    
polar_line(20, 1/3*np.pi)
```
